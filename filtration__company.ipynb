```python
import pandas as pd
```

```python
# Запишем считанные файлы в переменные df (список компаниий) и inn (список ИНН нужных компаний)
df = pd.read_excel('4_inn.xls')
inn = pd.read_csv('4_necessary_inn.txt', header=None)
```

```python
# Переведем даатфрейм в список через метод .to_list()
list_of_inn = inn[0].to_list()
list_of_inn
```

```python
# Отмеберм нужные нам компании через метожд query
df_companies = df.query('head_inn in @list_of_inn')
```

```python
# Считаем сумму доходов отфильтрованных компаний
df_companies['income,RUB'].sum()
```
